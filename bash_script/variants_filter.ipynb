```python
def read_vcf_with_header(vcf_file):
    """
    Read VCF file into pandas DataFrame and extract header lines.
    """
    vcf_header = []
    with open(vcf_file, "r") as f:
        for line in f:
            if line.startswith('#CHROM'):
                vcf_header.extend(line.strip().split('\t'))
    df = pd.read_csv(vcf_file, sep="\t", comment="#")
    df.columns = vcf_header
   
    return df


```

**CATEGORIZE**

- Case 1: both call same variants                                       True variants
- Case 2: GIAB call variants and  Sample do not                         Artifacts   
          Sample call variants and GIAB do not
- Case 3: GIAB calls and Sample also call variants but difference ALT   Artifacts

**TP, TN, FP, FN**
- Postive means calling variants
- Negative means not calling variants

- TP: both calls variants and they have the same REFs ALTs
- FN: sample does not call variant but GIAB does
- TN: Both do not call variants
- FP: sample calls variant but GIAB does not || they call variants at the same position but different ALTS

```python
m_df.head(150)
```

```python
###Define a custom function to determine positive and negative cases

def categorize_variants(df):
    if (df["ALT_GIAB"] == df["ALT_S"]) and (df["REF_GIAB"] == df["REF_S"]):
        return "TP"
    if (df["ALT_GIAB"] != df["ALT_S"]) and (df["REF_GIAB"] == df["REF_S"]):
        return "FP"
    if  df["ALT_GIAB"] and pd.isna(df["ALT_S"]):
        return "FP"
    if  pd.isna(df["ALT_GIAB"]) and df["ALT_S"]:
        return "FN"
    return "None"
```

```python
#Test
test_df = m_df[:150].copy()
# test_df
test_df['VAR_CATE'] = test_df.apply(categorize_variants, axis = 1)

test_df
```
