```python
import pandas as pd
m_df = pd.read_csv('/home/ndo/bash_output/variants_df/variants_df.csv', sep=",",encoding='utf-8')
```

```python
m_df
```

```python
#DROP the INDELS in GIAB
m_df.drop(m_df[m_df.VAR_CATE == 'None'].index, inplace=True)
```

```python
value_counts = m_df['VAR_CATE'].value_counts()

value_counts
```

```python
fn = m_df[m_df['VAR_CATE'] == "FN"]
fn = fn[['CHROM', 'POS']]
#change the name of pos column

fn = fn.rename(columns={"POS": "END"})

#add end column for the dataframe

fn['START'] = fn['END'] - 1

columns_titles = ["CHROM","START","END"]
fn=fn.reindex(columns=columns_titles)
fn
```

```python
#COVERT FN DATAFRAME TO BED FILE

# Convert the dataframe to a BED file
fn.to_csv('/home/ndo/bash_output/fn_bed/fn.bed', sep='\t', header=False, index=False)
```
